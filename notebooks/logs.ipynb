```python
import json
import os

def load_training_json(model_name: str, exp_name: str, seed) -> dict:
    file_path = os.path.join("logs", model_name, exp_name, str(seed), "training.json")
    with open(file_path, "r", encoding="utf-8") as f:
        data = json.load(f)
    return data

def load_all_training_logs(model_names, exp_name, seeds):
    all_logs = []
    for model in model_names:
        for seed in seeds:
            try:
                data = load_training_json(model, exp_name, seed)
                # 메타 정보 추가
                data["model_name"] = model
                data["seed"] = seed
                all_logs.append(data)
            except FileNotFoundError:
                print(f"[경고] 파일 없음: {model}/{exp_name}/{seed}")
            except json.JSONDecodeError:
                print(f"[경고] JSON 파싱 실패: {model}/{exp_name}/{seed}")
    return all_logs

bind_scores = []
site_scores = []

for i in range(1, 4):
    test = load_training_json('lstm', 'both_', i)
    first_key = list(test['final'].keys())[0]  # 보통 "best"나 "final" 같은 key일 것
    bind = test['final'][first_key]['scores']['bind']
    site = test['final'][first_key]['scores']['sites']
    
    bind_scores.append(bind)
    site_scores.append(site)

bind_df = pd.DataFrame(bind_scores)
site_df = pd.DataFrame(site_scores)

print("\n[BIND]")
print(bind_df.agg(['mean', 'std']).round(4))

print("\n[SITES]")
print(site_df.agg(['mean', 'std']).round(4))
```
